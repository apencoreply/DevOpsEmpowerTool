```python
import numpy as np
import matplotlib.pyplot as plt

# Data
categories = ['PostAPI', 'GetAPI', 'GetMean']
before = [11.9, 6.4, 2.8]
after = [6.3, 8.1, 6.0]

# Create a figure and a set of subplots
fig, ax = plt.subplots()

# Setting the positions
barWidth = 0.35
r1 = np.arange(len(before))
r2 = [x + barWidth for x in r1]

# Generate the bar chart
rects1 = ax.bar(r1, before, color='#7f6d5f', width=barWidth, edgecolor='grey', label='before')
rects2 = ax.bar(r2, after, color='#557f2d', width=barWidth, edgecolor='grey', label='after')

# Adding xticks
plt.xlabel('APIs', fontweight='bold')
plt.ylabel('Values', fontweight='bold')
plt.xticks([r + barWidth/2 for r in range(len(before))], categories)

# Set the title and labels
ax.set_title('Average Requests handled per second')

# Adding legend
plt.legend()

# Adding data labels
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
fig.tight_layout()

# Display the chart
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['PostAPI', 'GetAPI', 'GetMean']
before = [7, 88, 0]
after = [100, 0, 0]

# Create a figure and a set of subplots
fig, ax = plt.subplots()

# Setting the positions
barWidth = 0.35
r1 = np.arange(len(before))
r2 = [x + barWidth for x in r1]

# Generate the bar chart
ax.bar(r1, before, color='#1f77b4', width=barWidth, edgecolor='grey', label='Before Optimization')  # Blue
ax.bar(r2, after, color='#ff7f0e', width=barWidth, edgecolor='grey', label='After Optimization')  # Orange

# Adding xticks
plt.xlabel('APIs', fontweight='bold', fontsize=14)
plt.ylabel('Error Rate (%)', fontweight='bold', fontsize=14)
plt.xticks([r + barWidth/2 for r in range(len(before))], categories, fontsize=12)

# Set the title and labels
ax.set_title('Percentage of Errors in API Calls', fontsize=16)

# Adding legend
ax.legend(fontsize=12)

# Add grid
ax.grid(True)

# Add a tight layout
plt.tight_layout()

# Display the chart
plt.show()
```

```python
def myplot(title, y, before, after):
    import plotly.graph_objects as go

    # Data
    categories = ['PostAPI', 'GetAPI', 'GetMean']

    # Create traces
    fig = go.Figure()

    fig.add_trace(go.Bar(
        x=categories,
        y=before,
        name='Before',
        marker_color='rgb(55, 83, 109)'
    ))

    fig.add_trace(go.Bar(
        x=categories,
        y=after,
        name='After',
        marker_color='rgb(26, 118, 255)'
    ))

    # Customize layout
    fig.update_layout(
        title=title,
        titlefont_size=30,
        xaxis_tickfont_size=26,
        yaxis=dict(
            title=y,
            titlefont_size=30,
            tickfont_size=26,
        ),
        legend=dict(
            x=.9,
            y=1.0,
            font_size=24,
            bgcolor='rgba(255, 255, 255, 0)',
            bordercolor='rgba(255, 255, 255, 0)'
        ),
        barmode='group',
        bargap=0.15, # gap between bars of adjacent location coordinates.
        bargroupgap=0.1, # gap between bars of the same location coordinate.
        height=625  # adjust the height of the figure in pixels
    )

    fig.show()
```

```python
myplot('Time to handle 100 request (100x100)', 'Time (s)', [8, 16, 36], [16, 12, 17])
```

```python
myplot('Average Requests handeled per second', 'Avg throughput (req/s)', [11.9, 6.4, 2.8], [6.3, 8.1, 6.0])
```

```python
myplot('Percentage of Errors in API Calls', 'Error Rate (%)', [7, 88, 0], [100, 0, 0])
```
